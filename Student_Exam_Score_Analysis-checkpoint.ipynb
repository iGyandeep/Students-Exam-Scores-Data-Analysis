```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
#Change "WklyStudyHours" Columns
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5-Oct", "5-10")
df.head(5)
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution \n")
plt.show()
```

# Result:-
#From the above chart we have analysed that:
# -> The number of females in the data is more than the number of males

#Student average score based on the Parent Education

```python
gb_ParentEduc = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":"mean", "WritingScore":"mean"})
print(gb_ParentEduc)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb_ParentEduc, annot = True)
plt.title("Relationship between Parent's Education and Student's Score \n")
plt.show()
```

# ->Result
#From the above chart we have conclude that the education of the parents have a good impact on their child score.

```python
gb_ParentMaritalStatus = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gb_ParentMaritalStatus)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb_ParentMaritalStatus, annot = True)
plt.title("Reletionship between Parent's Marital Status and Student Score \n")
plt.show()
```

# Result:-
#from the above chart we concluded that there is no or negligible impact on the student's score due to their parent's marital status

```python
gb_PracticeSport = df.groupby("PracticeSport").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gb_PracticeSport)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb_PracticeSport, annot = True)
plt.title("Reletionship between Practice Sport and Student Score \n")
plt.show()
```

# Result:-
#from the above chart we concluded that there is no or negligible impact on the student's score due to their Sport's Practice
